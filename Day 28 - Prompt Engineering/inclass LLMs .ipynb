- Context in Generative AI (GenAI) refers to the **information and data that informs the AI's responses**
- It plays a crucial role in ensuring that AI models can provide relevant and accurate outputs by understanding the context of the data they are processing

**Context Window**: 
- A context window in artificial intelligence, particularly in large language models (LLMs), refers to 
    - **the amount of text (measured in tokens)** that the model can process, consider or **“remember”  at one time**
    - The context window is essentially the model's **working memory**, determining how much information it can consider when generating responses
   - 
   - This is also why, LLMs require GPUs to run
   - Parallel Proccessing to process those tokens at the same
   -
