### **Character Encoding** (ASCII and UTF-8)
- we have a way of telling computers that these are letters - `abcdef` or characters - `apiournaonrapnurjnaunbu*^$#$^%^(%#(%*%^&))`

- We have Boolean Logic, and Number Systems to help convert the `1010101010` the `364674` --- (`1010`)bin = (`10`)dec
- when you hit `A` ---> your computer sees `65`
- ASCII is an international standard of encoding characters
    - `A` = 65. `B` = 66
    - 7 bits for character encoding - used to be stored on 8-bit (1 byte) encoding scheme
    - 1 bit just there
    - _ _ _ _ _ _ _ _ = 2**7 = 128
    - this many characters were encoded
    - (A - z, a - z), (0 -9), basic punctuations
    - this did not support emojis, special characters, non english scripts
    - 
- Unicode Transformation Format - `UTF-8`
    - modern standard for web or python3
    - it is `variable width`
    - it uses 1 byte for standard english, which makes it compatible with ASCII as well
    - upto the first 4 bytes include all characters including emojis

- UTF-16 `2 to 4 bytes`, UTF-32 `fixed 4 bytes`

### TF-IDF (TERM FREQUENCY - Inverse Document Frequency)
- A statistical measure that evaluates the importance of a word in a document relative to a collection of documents(corpus)
- 
- It rewards the words that appear a lot in a specific `document` across the `corpus` ---> Term Frequency
- ![](https://media.geeksforgeeks.org/wp-content/uploads/20250207113146020518/The-TF-Formula.webp)
- it heavily penalizes words that appear in every document ---> Inverse Document Frequency
- penalizes more frequency
- ![](https://media.geeksforgeeks.org/wp-content/uploads/20250207113218469447/IDF-Formula.webp)
- Number of documents in the corpus
- 
- We are analyzing 1000 articles ---> corpus
    - `the` appears everywhere in all the corpus so tf-idf ---> score of 0
    - word `Jedi` appears 40 times in a specific article
    - tf-idf will give it a massive score to say that this word is very important for the document
    


```python
corpus = [
    "Forests are green and beautiful.",
    "Green trees make a forest beautiful"
]

from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd

tf_idf = TfidfVectorizer()

tf_idf_matrix = tf_idf.fit_transform(corpus)
# pd.DataFrame(tf_idf_matrix.toarray(), columns =tf_idf.get_feature_names_out())
# tf_idf.get_feature_names_out().shape

# multiplication of the tf idf in every place --> normalization ---> values exist 0 to 1
# 

idfs = tf_idf.idf_
words = tf_idf.get_feature_names_out()

# pd.DataFrame(idfs.T, columns= words)

word_idfs = (zip(words, idfs))

for i, x, in word_idfs: # unpacking zips
    print(i , x)
```

- **Document Similarity and Clustering**: By converting documents into numerical vectors TF-IDF enables comparison and grouping of related texts. This is valuable for clustering news articles, research papers or customer support tickets into meaningful categories.
- **Text Classification:** It helps in identify patterns in text for spam filtering, sentiment analysis and topic classification.
- **Keyword Extraction**: It ranks words by importance making it possible to automatically highlight key terms, generate document tags or create concise summaries.
- **Recommendation Systems**: Through comparison of textual descriptions TF-IDF supports suggesting related articles, videos or products enhancing user engagement.

Most popular word embedding models: 
### 1. **Word2Vec**
- learns to represent words as vectors in a way that captures their meaning based on context in the corpus
- Word2Vec looks at the whole corpus at once
- it is predictive
- it uses a shallow neural network to learn word meanings, by playing a guessing game with the context
- it does not know the context, just guesses
- you shall know a word by the company it keeps
    1. **Continous Bag of Words** 
    - predicts a **8** word based on the **surrounding** **context** words
    - give the model **context words** : the cat _____ the mat
    - model will try to predict the **target word**: ________sat_______
    - e.g. the cat *sits* on the mat
        - sits target --- the cat --- the mat --- context words


    2. **Skip Gram**
    - predicts **surrounding context words** from a **target word**
    - opposite of CBoW
    - give model **context words**, tries to predict the **target word**
    - 
![](https://swimm.io/wp-content/webp-express/webp-images/uploads/2023/11/word2vec--1024x559.png.webp)


- 
- it uses a neural network, we don't care about the output, we care about the weights of the hidden layer
- these **weights** are the **word embedding**
- 


```python
# pip install gensim
```

```python
from gensim.models import Word2Vec # anothe famous libarary for NLP
from nltk.tokenize import word_tokenize
import nltk

corpus = [
    "Machine learning is a subset of artificial intelligence",
    "Python is the most popular language for data science",
    "Neural networks are the backbone of deep learning",
    "Data science involves statistics and programming"
]

# we don't want to take the whole corpus, and turn it into tokens
token_sentences = [word_tokenize(sentence.lower()) for sentence in corpus]

# train word2vec
model = Word2Vec(
    sentences = token_sentences,
    vector_size = 100, # 100 numbers per word --- shape of the vector
    window = 5,        # how many words to look at 
    min_count = 1,      # ignore words that appear this many times
    workers = 4         # model uses 4 threads simulataneously -- 4 CPU cores
)
# computations will be small ---> there will be a lot of them 
# you have vectors of dimensions 100 - operations 
# parallel processing --- 
# the sweet spot will be between 3 to 12 workers --- the way the data moves, setting
# worker number as higher, might slow things down
# If you need reproducibility ---> workers = 1

# find words similar to any word --- python
similar_words = model.wv.most_similar('python')
similar_words

# actual vector of the word
vector = model.wv['python']
vector[:5]
```

- This vector ---> Dense vector ---> word embedding

### 2. **GloVe (Global Vectors for Word Representation)**

- GloVe focuses on global statistics of the corpus

- King - man + woman = queen

implementation is a little different from Word to Vec,

- 

```python
import gensim.downloader as api

model = api.load("glove-wiki-gigaword-50")
# dimensions === 50
```

```python
# King - man + woman = queen
result = model.most_similar(positive = ['king', 'woman'], negative= ['man'], topn=1)
result
```

```python
# check similarity scores
model.similarity('python', 'programming')

# find an odd one out
words = ['python', 'java', 'c++', 'broccoli']
words = ['python', 'java', 'c++', 'tomato', 'C', 'javascript']
model.doesnt_match(words)
```
