### Bag of Words
- is one of the simplest text representations
- ignores the order of words
- counts how many times a word appears in a document

```python
from sklearn.feature_extraction.text import CountVectorizer

corpus = [
    "Forests are green and beautiful.",
    "Green trees make a forest beautiful"
]

vectorizer = CountVectorizer()
bow_matrix = vectorizer.fit_transform(corpus)

bow_matrix.toarray()
vectorizer.get_feature_names_out()

import pandas as pd

df = pd.DataFrame(bow_matrix.toarray(), columns=vectorizer.get_feature_names_out())
df
```
