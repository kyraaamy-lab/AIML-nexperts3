### First step: an Image
- What is an image?
    - download an image into the work folder

To a computer, an image is just a massive `3D matrix` (if using python a NumPy array)
- Images are made of pixels ---> basic unit of an image ---> formed by combination of all the 3 layers of data
- the best language to use for processing such data is C++ ---> Fast
- Most fast libraries are written in C++ language
- The best library in python to use for computer vision --> OpenCV (open computer vision) 
- 
- Break an image down in python
- Two types of `images`:
    - **Grayscale Image**:
        - A 2D matrix ---> *(28 x 28)*
        - Each value is between ---> 0 to 255 ---> 8 bit images
        - Levels of intensity
        - 0 is black, 1 white

    - **Color Images**:
        - 3 separate layers
        - 3D matrix of stacked 2D matrices

A color image has **3 channels**: Red, Green, Blue.

| Color  | R   | G   | B   |
|--------|-----|-----|-----|
| Red    | 255 | 0   | 0   |
| Green  | 0   | 255 | 0   |
| Blue   | 0   | 0   | 255 |
| White  | 255 | 255 | 255 |
| Black  | 0   | 0   | 0   |
| Yellow | 255 | 255 | 0   |

```
Every pixel:
(r,b,g)
(0,0,0) ---> Black
(255,0,0) ---> Red 
(0,255,0) ---> Blue 
(0,0,255) ---> Green

(76,252,10) ---> bluish color`

#### To detect faces:
1. **Haar Cascade** 
    - a method that looks for specific patterns of light and dark in an image
    - For example:
        - the area around eyes is a bit darker than the nose bridge
2. Deep Learning approach ---> create a model that helps detect faces


````python
small_img = cv2.resize(img,(700,500), interpolation = cv2.INTER_NEAREST)

face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades +
                                      'haarcascade_frontalface_default.xml')

faces = face_cascade.detectMultiScale(small_img,
                                      scaleFactor = 1.1, # how much the image size is reduced
                                      minNeighbors = 5   # how many neighbors for each rectangle
                                      )

# [257, 190,  51,  51] - x, y, width, height

for (x,y, w, h) in faces:
    cv2.rectangle(small_img, (x, y), (x+w, y+h), (0,255,0), 2 )
    cv2.putText(small_img,'face', (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0,255,0), 2)

cv2.imshow("image", small_img) # it actually renders the image in a window
cv2.waitKey(5000) 
cv2.destroyAllWindows()
````
