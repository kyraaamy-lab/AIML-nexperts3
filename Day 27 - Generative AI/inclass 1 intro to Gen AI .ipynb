**2. Transformers**

- are DL models for NLP
- Foundation of modern GenAI and Large Language Models
- Transformer architecture power GPT (Generative Pretrained Transformer), and Gemini as well
- Enabled models to understand long range context in text
- It used the Self-Attention Mechanism - to weigh importance of every word relative to others in a sequence
- 
- Before Transformers we had RNNs and LSTMs
- RNNs had memory limits ---> 
- Recurrent Neural Networks they kept memory as well in each neuron
- 
- Transformers intuitively:
    - process the **entire text at once** instead of sequentially
    - use **Self-attention**
        - it is the mechanism AI uses to understand the context of a word based on the words surrounding it

- Transformers Include:
    - **Attention mechanisms**:
        - attention allows model to focus on different parts of input sequence
        - opposite to RNN that only processes data in order/sequence
    - **Parellel Processing**:
        - process all the words in a sentence simulatenously
        - faster and better results on long sequences of data
    - **Positional Encoding**:
        - they need a way to udnerstand the order of the words in a sentence
        - adds information about the position of each word in the sequence
        - this helps model understand the order of words

Examples: 
- BERT (Bidirectional Encoder Respresentation from Transformers)
    - transformer model trained to udnerstand the context of words in both directions (LtR or RtL)
    - question answering, sentiment analysis, and language understanding

- GPT (Generative PreTrained Transformer)
    - transformer model trained to generate text in conversational style
    - text completion, text summarization, creative writing 

Components:
1. Encoder
    - Takes an input sequence (sentence)
    - Processes it
    - Sequence of Vectors
    - Each word in teh sentence is converted to a vector using an embedding
2. Decoder
    - Takes the output from enccoder, generates the final prediction

**Text Summarization:**

- `facebook/bart-large-cnn` ---> 1.63 GB --> Createive summaries --- better at rephrasing things
- google-t5/t5-small -- 242 MB -- fast but simple summaries
- google-t5/t5-base --  890 MB -- middle ground accuracy and speed

```python
# Summarizer

summarizer = pipeline("summarization", model = "google-t5/t5-small" )
text = """
Forests are nature’s living cathedrals, where towering trees weave a green canopy that hums with life. 
Beneath their shade, the air feels cooler, richer, and alive with the scent of earth and leaves. 
Sunlight filters through in golden shards, dancing on mossy ground and whispering streams. 
Every rustle in the undergrowth hints at hidden creatures—birds calling from above, 
insects buzzing in the air, and animals moving quietly in the shadows. Forests are not just collections 
of trees; they are breathing, evolving worlds that shelter countless species, 
store ancient secrets, and remind us of the delicate balance that sustains life on Earth.
"""

summarizer(text, max_length = 50, min_length = 10)
```

```python
from transformers import pipeline

# Load the summarization pipeline
summarizer = pipeline("summarization")

def summarize_text(text, max_length=150, min_length=30):
    """Summarizes a given text."""
    summary = summarizer(text, max_length=max_length, min_length=min_length, do_sample=False)[0]['summary_text']
    return summary

# Example text
text = """
The Amazon rainforest, often called the "lungs of the Earth," is a vast and vital ecosystem. 
Spanning across nine South American countries, it covers an area of approximately 2.1 million square miles. 
This immense rainforest is home to an incredible diversity of plant and animal life, including millions of insect species, 
thousands of bird and mammal species, and countless varieties of trees and other plants. 
The Amazon plays a crucial role in regulating the Earth's climate by absorbing vast amounts of carbon dioxide. 
However, deforestation, primarily driven by agriculture and logging, poses a significant threat to the rainforest. 
The destruction of the Amazon not only leads to biodiversity loss but also contributes to global warming. 
Conservation efforts are essential to protect this invaluable ecosystem for future generations.
"""

# Generate summary
summary = summarize_text(text)

# Print results
print("Original Text:")
print(text)
print("\nSummary:")
print(summary)
```
