**2. Transformers**

- are DL models for NLP
    - Foundation of modern GenAI and Large Language Models
    - Transformer architecture powers GPT (Generative Pretrained Transformers), and Gemini as well
    - Enabled models to understand **long range context** in text
        - It used the *Self-Attention Mechanism* - to weigh importance of every word relative to others in a sequence
- 
- Before Transformers we had `RNNs` and `LSTMs`
- RNNs had memory limits ---> 
- Recurrent Neural Networks they kept memory as well in each neuron
- 
- Transformers intuitively:
    - Process the **entire text at once** instead of sequentially
    - Use **Self-attention**
        - it is the mechanism AI uses to understand the context of a word based on the words surrounding it

- **Transformer** concept includes:
    1. **Attention mechanisms**:
        - attention allows model to *focus on different parts* of input sequence
        - instead of processes data in order/sequence like RNN
        - 
    2. **Parellel Processing**:
        - process all the words in a sentence simulatenously
        - Bigger context windows
        - faster and better results on long sequences of data as well
        - GPUs are needed
        - 
    3. **Positional Encoding**:
        - they need a way to understand the order of the words in a sentence
        - adds information about the position of each word in the sequence
        - this helps model understand the order of words
        - are taken within the embeddings
        - 

**Attention Mechanisms**
- helps a model focus on the most relevant parts of its input when generating an output
- Instead of treating every token equally, 
    - the model compares the **current representation** against all **other token representations**
    - assigns a score to each comparison

- 
- **Attention** is built around three components: **Queries**, **Keys**, and **Values**. QKV

    - **Queries**: 
        - The signal sent by the current token to "search" for relevant information in the rest of the sequence

    - **Keys**: 
        - The identifier for each token in the sequence that determines *how well it matches the search signal*
        
    - **Values**: 
        - The actual informational content that is retrieved and used when a match between a **Query and a Key** is found.

### **Named Entity Recognition (NER)**

- is an NLP technique that identifies and classifies key entities in text, such as people, organizations, locations, dates, and more, transforming unstructured text into structured data.
- The Smart Highlighter
Imagine you have to read a 500-page legal document, and your boss tells you to highlight every single Person's name in yellow, every Company in green, and every City in blue. Doing that manually would take hours

- If you feed an NER model this sentence:

"Tim Cook announced that Apple will open a new store in Paris."

The model will output:
- Tim Cook -> PERSON
- Apple -> ORGANIZATION
- Paris -> LOCATION

- `elastic/distilbert-base-cased-finetuned-conll03-english` - 260 MB
- `dslim/bert-base-NER` - 430 MB
- `boltuix/bert-mini` -  15 to 50 MB

```python
from transformers import pipeline

ner_highlighter = pipeline("ner", 
                           aggregation_strategy="simple", # We use aggregation_strategy="simple" to keep multi-word names together
                           model = 'dslim/bert-base-NER'
                           )
```

```python
# 2. Provide a sentence
text = "Elon Musk founded SpaceX in Hawthorne, California."

print("Scanning text for entities...\n")

# 3. Run the model!
results = ner_highlighter(text)

results
```

```python
# 4. Print the results clearly
for entity in results:
    word = entity['word']
    label = entity['entity_group']
    confidence = round(entity['score'] * 100, 2)
    
    print(f"Found '{word}' -> Category: {label} (Confidence: {confidence}%)")
```

### **Text Generation Locally**

- LLMs Locally - You downloaded a model on your system, and are using it offline
- 

- generate text from prompts
- on a local model
- 
- gpt2 - 550MB - 
- Qwen/Qwen-0.6B - 1.2 GB - 
- microsoft/phi-4-mini - 7.5 GB - 
- meta-llama/Llama-3.1-8B - 15GB - needs a good GPU


```python
from transformers import pipeline

generator = pipeline("text-generation", model = "gpt2")

text = """
Who is Muamar Gaddafi?
"""

result = generator(text,
                   #max_new_tokens = 100, # how many new words to write
                   temperature = 0.5, # risk setting ---- 0.1 --> safe boring, repetitive higher is more chaotic
                   # o.7 for balanced # 1.2 for more wild/unpredictible
                   do_sample = True # tells the model to randomly choose words based on probaility -- not most likely everytime
                   )
print(result[0]['generated_text'])
```
