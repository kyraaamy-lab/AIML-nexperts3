5. **Stemming/Lemmatization:**
- Find the root of the words
    - Reduce the words to their roots
    - "jumping" ---> "jump"
Result: ["quick", "brown", "fox","jump"]

- e.g run running runner ---> run run run

- 
- Affixes - parts that are added to a base word (root)
    - prefix
        - `un`:  do ---> undo 
    - suffix
        - `ing`: run ---> running 

- infixes - words within a word
    - a bit more rare, 
    - e.g. father-in-law

- circumfixes - words that go around a word
    - eg. `en`:  light ----> enlighten

**Stemming**
- it will help in text analysis, by treating different forms of the word as same or equivalent
    - removes the affixes to get the base word
    - sometimes it produces words that do not make sense
- running -> run
- happier -> happi
- quickly -> quick
- 
- different words will be taken as equivalents
- chang`ing` -> chang
- chang`ed` -> chang
- chang`e` -> chang
- 

- studies -> studi
- stud`ying`-> studi
- study -> studi

### Preprocessing Pipeline at once - NLTK

```python
text = """
Forests are nature’s living cathedrals, where towering trees weave a green canopy that hums with life. 
Beneath their shade, the air feels cooler, richer, and alive with the scent of earth and leaves. 
Sunlight filters through in golden shards, dancing on mossy ground and whispering streams. 
Every rustle in the undergrowth hints at hidden creatures—birds calling from above, 
insects buzzing in the air, and animals moving quietly in the shadows. Forests are not just collections 
of trees; they are breathing, evolving worlds that shelter countless species, 
store ancient secrets, and remind us of the delicate balance that sustains life on Earth.
"""
```

```python
import nltk
import string

from nltk.tokenize import sent_tokenize, word_tokenize, WhitespaceTokenizer
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer

# 1. Tokenization
# nltk.download('punkt') # might be needed for tokenization

sentences = sent_tokenize(text)
print("# of Sentences: ",len(sentences))

words = word_tokenize(text) # punctuation aware tokenization
print("# of words: ",len(words))

char_tokens = list(text)
print("# of tokens ", len(char_tokens))

# whitespace tokenization
whitespace_tokens = text.split()
print("# of whitespace_tokens ", len(whitespace_tokens))

# stop word removal
stopwords = stopwords.words('english')

filter_stopwords = [word for word in words if word not in stopwords]
print("# after filter_stopwords ", len(filter_stopwords))

# remove punctuation
string.punctuation # !"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~

filter_punctuations = [word for word in words if word not in string.punctuation]
print("# after filter_punctuations ", len(filter_punctuations))

# ---------------------------
print("--"* 20)

# stemming and lemmatization

new_text = """
Forests are nature's living where towering - studying running going runner improvement improving
"""

new_words = word_tokenize(new_text)

stemmer = PorterStemmer()

lemmatizer = WordNetLemmatizer()

stemmed_words = [stemmer.stem(word) for word in new_words]
print("stemmed words: ", stemmed_words)

lemmatized_words = [lemmatizer.lemmatize(word, pos= 'v') for word in new_words] # POS tag - pos = 'v' --> remove the -ing
print("Lemmatized words: ", lemmatized_words)
```
