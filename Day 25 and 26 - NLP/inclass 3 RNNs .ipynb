## $$Recurrent-Neural-Networks$$

- [RNN](https://www.geeksforgeeks.org/machine-learning/introduction-to-recurrent-neural-network/)

Recurrent Neural Networks (RNNs) are a class of neural networks designed to process **sequential data** by **retaining information** from previous steps
- Temporal data is information associated with time - e.g. Time Series data - 
- Sequential data like Text data 
- 
- retaining previous information means we can retain some context of what came before
- 
- In CNNs etc you predict output class or a value
- In RNNs we try to predict the next word
-
![](https://media.geeksforgeeks.org/wp-content/uploads/20251215085936469976/introduction_to_recurrent_neural_network.webp)

- Imagine reading a sentence and you try to predict the next word
- RNNs work similarly by “remembering” past information i.e it considers all the earlier words to choose the most likely next word.
- This memory of previous steps helps the network understand context and make better predictions.

Recurrent Neurons

- The fundamental processing unit in RNN is a Recurrent Unit. 
- They hold a hidden state that maintains information about previous inputs in a sequence. 
- Recurrent units can "remember" information from prior steps by feeding back their hidden state, allowing them to capture dependencies across time.

![](https://media.geeksforgeeks.org/wp-content/uploads/20251215090149060237/nfa-.webp)

RNN Unfolding

- unrolling enables backpropagation through time (BPTT) 
- a learning process where errors are propagated across **time steps** 
    - to adjust the network’s weights
- 
![](https://www.researchgate.net/profile/Abdelhakim-Dorbane-2/publication/374206433/figure/fig1/AS:11431281200368786@1697879902068/A-schematic-illustration-of-an-unrolled-RNN-architecture-This-unrolling-visually.ppm)

Recurrent Neural Network Architecture

- RNNs share similarities in input and output structures
- with other deep learning architectures but differ significantly in 
    - how information flows from input to output
    - 
- Unlike traditional DNNs where each dense layer has **distinct weight matrices**

- RNNs use **shared weights** across **time steps**, allowing them to remember information over sequences
    - at every time step there will be a weight matrix
    - all matrices are shared by a layer, across time
