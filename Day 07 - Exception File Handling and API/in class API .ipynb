### Dog CEO's Dog API
- 
- a website, which we will access through requests package


```python
import requests
from IPython.display import Image

url = "https://dog.ceo/api/breeds/image/random"

response = requests.get(url)

status_code = response.status_code
status_code
```

HTTP Status Codes: (error codes)
- 200 - The request was successful
- 404 - Not Found - The server could not find the requested page
- 500 - Internal Server Error - processing request, but faced an error

```python
data = response.json() # gives us a message and a status

type(data)
```

```python
if status_code == 200:
    response = requests.get(url)
    data = response.json()
    dog_image_url = data['message'] 

    print("here is the image of a dog!", dog_image_url)
else:
    print("Failed to fetch data")

Image(url= dog_image_url,
      width = 300,
      height= 400)
```

### **Web Scraping**

- going to websites and getting data out

```python

import json

url = "https://thispersondoesnotexist.com/?random"

response = requests.get(url)

string = response.headers['Report-To']

url = json.loads(string)

image_url = url['endpoints'][0]['url']
image_url
```

```python
Image(url= image_url)
```

```python
import requests
import time
from IPython.display import Image, display

url = "https://thispersondoesnotexist.com/image"

# This specific header combo is usually required to stop the 404 error
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36",
    "Referer": "https://thispersondoesnotexist.com/"
}

# Create a session to look more like a real user
session = requests.Session()

for i in range(3):
    response = session.get(url, headers=headers)
    
    if response.status_code == 200:
        print(f"Image {i+1}:")
        display(Image(response.content))
    else:
        print(f"Failed: Status {response.status_code}. Try running the cell again.")
        
    time.sleep(2) # Don't remove this! The site will block you if you go too fast.
```
